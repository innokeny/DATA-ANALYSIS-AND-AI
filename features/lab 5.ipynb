```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('/content/drive/MyDrive/Python/факультатив/LB5/train.csv') 
df['Cabin'] = df['Cabin'].fillna('M')
df
```

```python
Deck = []
for i in range(len(df.axes[0])):
  Deck.append(df['Cabin'][i][0])
df = df.assign(Deck = Deck)
df
```

```python
df = df.assign(Binned_Age = pd.qcut(df['Age'], q = 10))
df
```

```python
df = df.assign(Binned_fare = pd.qcut(df['Fare'], q = 13))
df
```

```python
temp = df.SibSp + df.Parch + 1
for i in range(len(temp)):
  if (temp[i] == 1): temp[i] = 'alone'
  if (temp[i] == 2 or temp[i] == 3 or temp[i] == 4): temp[i] = 'small'
  if (temp[i] == 5 or temp[i] == 6): temp[i] = 'medium'
  if (temp[i] == 7 or temp[i] == 8 or temp[i] == 11): temp[i] = 'large'
df = df.assign(Family_Size = temp)
df
```

```python
df = df.assign(Ticket_Frequency = df.groupby('Ticket')['Ticket'].transform('count'))
df
```

```python
df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
df
```

```python
df['Is_Married'] = 0
df['Is_Married'].loc[df['Title'] == 'Mrs'] = 1
df
```

```python
df['Title'] = df['Title'].replace(['Miss', 'Mrs','Ms', 'Mlle', 'Lady', 'Mme', 'the Countess', 'Dona'], 'Miss/Mrs/Ms')
df['Title'] = df['Title'].replace(['Dr', 'Col', 'Major', 'Jonkheer', 'Capt', 'Sir', 'Don', 'Rev'], 'Dr/Military/Noble/Clergy')
df
```

```python
data = pd.read_csv('/content/drive/MyDrive/Python/факультатив/LB4/train.csv')
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'object']
numerical_columns = [c for c in data.columns if data[c].dtype.name != 'object']
data = data.fillna(data.median(axis=0), axis=0)
data_describe = data.describe(include=[object])
for c in categorical_columns:
  data[c] = data[c].fillna(data_describe[c]['top'])
binary_columns = [c for c in categorical_columns if
data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if
data_describe[c]['unique'] > 2]
for c in binary_columns:
  top = data_describe[c]['top']
  top_items = data[c] == top
  data.loc[top_items, c] = 0
  data.loc[np.logical_not(top_items), c] = 1
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
data_numerical = data[numerical_columns]
data_numerical = (data_numerical - data_numerical.mean()) / data_numerical.std()
data = pd.concat((data_numerical, data[binary_columns], data_nonbinary), axis=1)
data = pd.DataFrame(data, dtype=float)
```

```python
X_train = data.iloc[:,2:]
#y_train = data['Survived'].astype('category')
y_train = df['Survived'].values
```

```python
y_train.dtype
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(random_state=0)
forest.fit(X_train, y_train)
importances = forest.feature_importances_

forest_importances = pd.Series(importances)

std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

```python
from sklearn.inspection import permutation_importance

model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)
r = permutation_importance(model, X_train, y_train,
                            n_repeats=2,
                            random_state=0)
```

```python
from sklearn.base import clone

def dropcol_importances(rf, X_train, y_train):
  rf_ = clone(rf)
  rf_.random_state = 999
  rf_.fit(X_train, y_train)
  baseline = rf_.oob_score
  imp = []
  for col in X_train.columns:
    X = X_train.drop(col, axis=1)
    rf_ = clone(rf)
    rf_.random_state = 999
    rf_.fit(X, y_train)
    o = rf_.oob_score_
    imp.append(baseline - o)
  imp = np.array(imp)
  I = pd.DataFrame(
    data={'Feature':X_train.columns,
      'Importance':imp})
  I = I.set_index('Feature')
  I = I.sort_values('Importance', ascending=True)
  return I

model = RandomForestClassifier(random_state=0,warm_start=True, bootstrap=True)
model.fit(X_train, y_train)
imp = dropcol_importances(model, X_train, y_train)
print(imp)
```
